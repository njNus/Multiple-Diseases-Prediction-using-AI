Now we have split our data into training and test sets, it's time to build ML model

We'll train it on (find the patterns) on the training data

And we'll test it (use the patterns) on the test set

We are going to implent the model

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
X=df.iloc[:,0:13]
y=df.iloc[:,13:14]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)

ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras import regularizers

#Creating a pipeline
model = Sequential()

#1st hidden layer with input layer
model.add(Dense(units=145,activation="relu",input_dim=13, kernel_regularizer=regularizers.l2(0.01)))

#2nd hidden layer
model.add(Dense(units=120,activation="relu",kernel_regularizer=regularizers.l2(0.01)))

#3rd hidden layer
model.add(Dense(units=70,activation="relu",kernel_regularizer=regularizers.l2(0.01)))



#output layer
model.add(Dense(units=1,activation="sigmoid"))

model.summary()
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
model_his=model.fit(X_train,y_train,validation_split=0.30, batch_size=55,epochs=200,verbose=1)


y_pred=model.predict(X_test)
y_pred = (y_pred > 0.45)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
print(classification_report(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
# summarize history for accuracy
plt.plot(model_his.history['accuracy'])
plt.plot(model_his.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

```

```python
# summarize history for loss
plt.plot(model_his.history['loss'])
plt.plot(model_his.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
